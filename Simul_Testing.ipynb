```python
import numpy as np
from SmartTruck import SmartTruck
from kalmanfilter2 import KalmanFilter
from simulation import Simulation
```

```python
ds = 2 # dimensions
t = 10 # number of time steps to run
dt = 0.1 # time step length
ep_mag = 10 # process noise variation
ep_dir = np.array([0.5])
nu = 0.4 # measurement noise variation (both position and velocity)
initial = np.ones(ds*2)
initial.shape = (ds*2,1)

gen = SmartTruck(initial, t, dt, ep_mag, ep_dir, nu)
sim = Simulation(gen, KalmanFilter, seed_value=1)
```

```python
sim.generate()

test = np.ones(ds*2)*5
test.shape = (ds*2,1)

sim.predict(x0 = test)
```

```python
sim.plot()
```
