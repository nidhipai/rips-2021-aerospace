## Example: How to run an MHT simulation in a Jupyter Notebook ##

```python
# Import mtt package and other necessary packages
import mtt
import numpy as np
import matplotlib.pyplot as plt

# Set default simulation values
obj1 = np.array([[0], [0], [1], [1]])
obj2 = np.array([[-5], [5], [1], [1]])

initial = {0: obj1, 1:obj2}
dt = 1
ep_normal = .001
ep_tangent = .001
nu = 1
ts = 30
miss_p = 0
lam = 2
fa_scale = 2

x_lim = 40
y_lim = 40
new_obj_prop = 0.00

# Set up generator and tracker objects to generate the data
gen = mtt.MultiObjFixed(initial, dt, ep_tangent, ep_normal, nu, miss_p, lam = lam, fa_scale = fa_scale, x_lim = x_lim, y_lim = y_lim, new_obj_prop = new_obj_prop)
params = gen.get_params()

tracker = mtt.Presets.standardMHT(params, miss_p, lam, gate_size=0.95, gate_expand_size=0.5, tot=0.001, tmm=0.1, tnt=0.6, prune_time=3, starting_pos = initial)
sim = mtt.Simulation(gen, tracker, seed_value = 7341818)

# Run the simulation and get a prediction from the object trajectory
sim.generate(ts)
sim.predict()
```

```python
# Plot the trajectories (may want to adjust font sizes if the plot is resized on your computer)
sim.plot();
```

```python
# Can access attributes of the simulation like so:
print(sim.mota)
print(sim.motp)
```
