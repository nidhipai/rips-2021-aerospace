```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mean = (0, 0)
cov = [[1, 0], [0, 1]]
X = np.random.multivariate_normal(mean, cov, 10000)
plt.plot(X[:,0], X[:,1], marker = '.', linewidth = 0)
axes=plt.gca()
axes.set_aspect(1)
```

```python
mean = (0, 0)
cov = [[1, 0.5], [0.5, 1]]
X = np.random.multivariate_normal(mean, cov, 10000)
plt.plot(X[:,0], X[:,1], marker = '.', linewidth = 0)
axes=plt.gca()
axes.set_aspect(1)
```

```python
theta = math.pi / 2
c = math.cos(theta)
s = math.sin(theta)
rotation = np.array([[c, -s], [s, c]])
mean = (0, 0)
cov = np.array([[1, 0.5], [0.5, 1]])
rotated_cov = rotation @ cov @ rotation.T
X = np.random.multivariate_normal(mean, rotated_cov, 10000)
plt.plot(X[:,0], X[:,1], marker = '.', linewidth = 0)
axes=plt.gca()
axes.set_aspect(1)
```
